```python
# Write a program to find the number of vowels, consonants, digits, and white
# space characters in a string.

a = input('Enter the string')
vowels_cha= 'aeiouAEIOU'

vowels = 0
string = 0
digits = 0
white_space = 0

for i in a:
    if i in vowels_cha:
        vowels+=1
    elif i.isalpha():
        string+=1
    elif i.isdigit():
        digits+=1
    elif i.isspace():
        white_space +=1
print('vowels',vowels)
print('consonants',string)
print('digits',digits)
print('white_space',white_space)        
```

```python
# Write a Python function to reverses a string if its length is a multiple of 5
def reverse(s):

    if len(s) % 5 == 0:
        return s[::-1]
    return s

input_string = "HelloWorld"
result = reverse(input_string)
print(result) 
```

```python
import numpy as np

my_array = np.array([1, 2, np.nan, 4, np.nan, 6])

# Create a mask for NaN values
mask = np.isnan(my_array)

# Use the inverse of the mask to filter out missing values
filtered_array = my_array[~mask]

print("Original Array:", my_array)
print("Filtered Array (without NaNs):", filtered_array)
```

```python
import numpy as np

my_array = np.array([1, 2, np.nan, 4, np.nan, 6])

# Create a mask for NaN values
mask = np.isnan(my_array)

# Use the inverse of the mask to filter out missing values
filtered_array = my_array[mask]

print("Original Array:", my_array)
print("Filtered Array (without NaNs):", filtered_array)
```
