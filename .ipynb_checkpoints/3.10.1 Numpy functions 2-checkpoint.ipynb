# sort()

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore') 
```

```python
ran_nums = np.random.randint(1,100,25)
ran_nums
```

## Ascending

```python
np.sort(ran_nums)
```

```python
sorted_ran_nums=sorted(ran_nums,reverse=True)
sorted_ran_nums
```

## Descending

```python
np.sort(ran_nums)[::-1]
```

```python
desc_ran_nums_sorted = np.array(sorted(ran_nums,reverse=True))
desc_ran_nums_sorted
```

## sorting of matrix

```python
np.sort([2,3,0]) 
```

### sorting column wise

```python
np.sort([[2,3,0],
         [7,5,8]],axis=1) 
```

```python
np.sort([[2,3,0],
         [7,5,8]],axis=1)
```

```python
a = np.array([[1,4], 
              [3,1]])
a.sort(axis=1)
print(a) 
```

### sorting row wise

```python
matrix = [[8,3,9],
          [1,5,0]]
np.sort(matrix,axis=0)
```

```python
a = np.array([[1, 4],[1, 3]])
a.sort(axis=0)
print(a)
```

# replacing any item in any array using indexing

```python
array = np.array([16,17,2,5,7])
array[1]=100
array
```

```python
array = np.array([[21,16,17],
                  [2,5,3],
                  [7,0,8]])
array[0][0]=100
array 
```

```python
array = np.array([[21,16,17],
                  [2,5,3],
                  [7,0,8]])
print(array[0:2,0:2])
array[0:2,0:2]=[[100,200],[300,400]]
array 
```

# Inserting item into Array

```python
inserts item at any specific index into the array

Syntax:
np.insert(array_name,index,value) 
```

```python
array = np.array([16,17,2,5,7])
np.insert(array,1,100) 
```

```python
array = np.array([16,17,2,5,7])
np.insert(array,-1,1000)
```

```python
array = np.array([16,17,2,5,7])
np.insert(array,len(array),1000)
```

```python
array = np.array([[21,16,17],
                  [2,5,3],
                  [7,0,8]])
np.insert(array,[0][0],100) 
```

# Concatenate()

```python
import numpy as np 
```

```python
array1 = np.array([2,3,4])
array2 = np.array([5,3,4])

final_array = np.concatenate([array1,array2])
final_array
```

```python
array1 = np.array([2,3,4])
array2 = np.array([5,3,4])
array3 = np.array([5,3,9])


final_array = np.concatenate([array1,array2,array3])
final_array
```

```python
array1 = np.array([2,3,4])
array2 = np.array([5,3,4])
array3 = np.array([5,3,9])


final_array = np.concatenate([array1,array2,array3])
final_array.reshape(3,3)
```

## Concatenation along rows in matrix

```python
array1 = np.array([[2,3,4],
                   [5,6,7]])

array2 = np.array([[5,8,9]])

np.concatenate([array1,array2],axis=0)
```

```python
array1 = np.array([[2,3,4],
                   [5,6,7]])

array2 = np.array([[5,8,9]])
array3 = np.array([[0,0,0]])

np.concatenate([array1,array2,array3],axis=0) 
```

## Concatenation along columns in matrix


```python
array1 = np.array([[2,3,4],
                   [5,6,7]])

array2 = np.array([[5],
                   [0]])

np.concatenate([array1,array2],axis=1)
```

```python
array1 = np.array([[2,3,4],
                   [5,6,7]])

array2 = np.array([[5,5],
                   [0,6]])
array3 = np.array([[1],
                   [1]])

con_c_array=np.concatenate([array1,array2,array3],axis=1)
con_c_array
```

# Looping on array

```python
array1 = np.array([2,3,4,6,7])
for i in array1:
    print(i) 
```

```python
array1 = np.array([[2,3,4,6,7]])
for i in array1:
    print(i) 
```

```python
array1 = np.array([2,3,4,6,7])
for i in enumerate(array1):
    print(i)
```

```python
array1 = np.array([2,3,4,6,7])
for index,value in enumerate(array1):
    print(index,value)
```

```python
RRR = np.array([[2, 3, 4],
       [5, 6, 7],
       [5, 8, 9],
       [0, 0, 0]])

for i in RRR:
    print(i)
    
print()
for i,v in enumerate(RRR):
    print(i,v)
```

## np.nditer()

```python
RRR = np.array([[2, 3, 4],
       [5, 6, 7],
       [5, 8, 9],
       [0, 0, 0]])

for i in np.nditer(RRR):
    print(i)
```

```python
for i in (RRR): #[2,3,4] ,[5,6,7]
    for j in range(len(i)): #[0,1,2]
        print(i[j]) 
```

## np.ndenumerate()

```python
display(RRR)
for i,v in np.ndenumerate(RRR):
    print(i,v)
```
